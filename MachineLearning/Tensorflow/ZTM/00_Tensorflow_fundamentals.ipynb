```python
# check number of dimensions
tensor.ndim
```

What we've created so far:

* Scalar: a single number
* Vector: a number with direction (e.g. wind speed and direction)
* Matrix: a 2-dimensional array of numbers
* Tensor: an n-dimensional array of numbers (when n can be any number, a 0-dimensional tensor is a scalar, a 1-dimensional tensor is a vector)

### Creating random tensors

Random tensors are tensors of some abitary size which contain random numbers

```python
# Create two random (but the same) tensors
random_1 = tf.random.Generator.from_seed(42) # Set seed for reproducibility
random_1 = random_1.normal(shape=(3,2))
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3,2))

# Are they equal
random_1, random_2, random_1 == random_2
```

### Shuffle the order of elements in a tensor

```python
# Shuffle a tensor (valuAble for when you want o shuffle your data so the inherent order doesn't effect learning)
not_shuffled = tf.constant([[10,7], [3,4], [2,5]])
not_shuffled.ndim
```

```python
not_shuffled
```

```python
# shuffle non-shffuled tensor
tf.random.set_seed(42) # Global level random seed
tf.random.shuffle(not_shuffled, seed=42) # Operation level random seed
```

Read through TensorFlow documentation on random seed generation: https://www.tensorflow.org/api_docs/python/tf/random/set_seed

It looks like if we want our shuffled tensors to be in the same order, we've got ot use the global level operation seed in cojunction to operation seed

### Getting information from tensors

When dealing with tensors you probably want to be aware of the following attributes:
* Shape
* Rank
* Axis or dimension
* Size

```python
# Create a rank 4 tensor (4 dimensions)
rank_4_tensor = tf.zeros(shape=[2,3,4,5])
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
# Get various attributes of our tensor
print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dimensions (rank):", rank_4_tensor.ndim)
print("Shape of tensor: ", rank_4_tensor.shape)
print("Elements along the 0 axis:", rank_4_tensor.shape[0])
print("Elements along the last axis: ", rank_4_tensor.shape[-1])
print("Total number of elements in our tensor", tf.size(rank_4_tensor).numpy())
```

### Indexing tensors

Tensors can be indexed just like Python lists.

```python
# Get the first 2 elements of each dimension
rank_4_tensor[:2, :2, :2, :2]
```

```python
# Get the first element from each dimension except for the final one
rank_4_tensor[:1,:1,:1, :]
```

```python
# Create a rank 2 tensor (2 dimensions)
rank_2_tensor = tf.constant([[10,7],[3,4]])
rank_2_tensor.shape, rank_2_tensor.ndim
```

```python
# Get the last item of each of our rank 2 tensor
rank_2_tensor[:,-1]
```

```python
# Add in extra dimension to our rank 2 tensor
rank_3_tensor = rank_2_tensor[..., tf.newaxis] # ... means [:,:]
rank_3_tensor
```

```python
# Alternative to tf.newaxis
tf.expand_dims(rank_2_tensor, axis=-1) # -1 means expand the final index
```

```python
# Expand the 0-axis
tf.expand_dims(rank_2_tensor, axis=0)
```

**Matrix Multiplication**

In machine learning, matrix muliplication is one of the most common operations

There are two rules our tensors (or matrices) need to fulfil if we're going to matrix multiply them:
1. Then inner dimensions must match
2. The resulting matrix has the shape of outer dimenesions

```python
# Matrix multiplication with tensorflow
tf.matmul(tensor, tensor) # alias for tf.linalg.matmul
```

```python
# Matrix multiplication with Python operator `@`
tensor @ tensor
```

```python
# Create a tensor (3, 2) 
X = tf.constant([[1,2],
                 [3,4],
                 [5,6]])
# Create another (3, 2)
Y = tf.constant([[7,8],
                 [9, 10],
                 [11, 12]])
X,Y
```

```python
tf.transpose(X)
```

```python
# Try to matrix multiply tensors of same shape
tf.transpose(X) @ Y
```

```python
# Can do the same with reshape
tf.reshape(Y, shape=(2,3))
```

```python
X @ tf.reshape(Y, shape=(2,3))
```

```python
X, tf.transpose(X), tf.reshape(X, shape=(2,3))
```

**The dot product**

Matrix multiplication is also referred to as the dot product

We can perform matrix multiplication using:
* `tf.matmul()`
* `tf.tensordot()`

```python
# perform the dot product on X and Y (requires X or Y to be transposed)
tf.tensordot(tf.transpose(X), Y, axes=1)
```

Generally, when performing matrix multiplication on two tensors and one of the axes doesn't line up, you will tranpose (rather than reshape) one of the tensors to get satisify the matrix multiplication rules.

### Aggregating tensors

Aggregating tensors = condesing them from multiple values down to a smaller amount of values.

```python
# Get the absolute values
D = tf.constant([-7, -10])
D
```

```python
# Get the absolute values
tf.abs(D)
```

Let's go through the following forms of aggregation:
* Maximum
* Minimum
* mean
* sum


```python
# Create a random tensor with values between 0 and 100 of size 50
E = tf.constant(np.random.randint(0, 100, size=50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# Minimum
tf.reduce_min(E)
```

```python
# Maximum
tf.reduce_max(E)
```

```python
# mean
tf.reduce_mean(E)
```

```python
# Sum
tf.reduce_sum(E)
```

```python
# Variance
tf.math.reduce_variance(tf.cast(E, dtype=tf.float32))
```

```python
# Variance using tensorflow probability package
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
# STD (Standard Deviation)
tf.math.reduce_std(tf.cast(E, dtype=tf.float32))
```
